```python
from plotly.offline import iplot
import plotly as py
import plotly.tools as tls
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn import datasets
import pandas as pd
import plotly.express as px
```

```python
import cufflinks as cf
```

```python
py.offline.init_notebook_mode(connected = True) 
cf.go_offline()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaled_iris = iris.drop('species', axis = 1)
scaled_data = scaler.fit_transform(scaled_iris)
```

```python
pca = PCA(n_components = 3)
iris_pca = pca.fit_transform(scaled_data)
```

```python
iris_pca.shape
```

```python
iris_data_new = pd.DataFrame(iris_pca)
```

```python
iris_data_new.head()
```

```python
iris_final = pd.concat([iris_data_new,iris['species']])
```

```python
iris_final.columns = ['1st PCA','2nd PCA','3rd PCA']
iris_final['species'] = iris['species']
```

```python
iris_final.head()
```

```python
px.scatter_3d(iris_final, x = '1st PCA', y = '2nd PCA', z = '3rd PCA', color = 'species')
```
